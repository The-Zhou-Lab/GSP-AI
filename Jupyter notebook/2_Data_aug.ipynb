```python
#正对原始数据augment
#针对 Huang
import os, random, shutil
import cv2
from skimage import io
from skimage.io import imread
import numpy as np
from skimage.color import rgb2yuv, rgb2hsv, rgb2gray, yuv2rgb, hsv2rgb
from scipy.signal import convolve2d

#sharpening = 0.4    
sharpen = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]])
def multi_convolver(image, kernel, iterations):
    for i in range(iterations):
        image = convolve2d(image, kernel, 'same', boundary = 'fill',
                           fillvalue = 0)
    return image
def sharpening(fileDir, kernel,tarDir,rate,iterations = 1):
    pathDir = os.listdir(fileDir);
    filenumber = len(pathDir)
    picknumber=int(filenumber*rate) #按照rate比例从文件夹中取一定数量图片
    sample = random.sample(pathDir, picknumber)  #随机选取picknumber数量的样本图片
    for name in sample:
        img_path = fileDir+name
        image = imread(img_path)
        img_hsv = rgb2hsv(image)   
        img_hsv[:,:,2] = multi_convolver(img_hsv[:,:,2], kernel, iterations)
        final_image_hsv = hsv2rgb(img_hsv)
        new_img = final_image_hsv
        b,g,r  = cv2.split(final_image_hsv)
        img2 = cv2.merge([r,g,b])
        path_save = tarDir+name.split('.')[0]+'s.png'
        print(path_save)
        cv2.imwrite(path_save,img2*255,[int(cv2.IMWRITE_PNG_COMPRESSION), 0])
#Increase 
def Increase(fileDir,tarDir,rate):
    pathDir = os.listdir(fileDir);
    filenumber = len(pathDir)
    picknumber=int(filenumber*rate) #按照rate比例从文件夹中取一定数量图片
    sample = random.sample(pathDir, picknumber)  #随机选取picknumber数量的样本图片
    for name in sample:
        img_path = fileDir+name
        image = imread(img_path)
        new_image = np.zeros(image.shape, image.dtype)
        alpha = 1.1 # Simple contrast control
        beta = 1.1  # Simple brightness control
        y = image[:,:,0];x = image[:,:,1];c = image[:,:,2]
        new_image[:,:,0] = np.clip(alpha*image[:,:,0]+beta,0,255)
        new_image[:,:,1] = np.clip(alpha*image[:,:,1]+beta,0,255)
        new_image[:,:,2] = np.clip(alpha*image[:,:,2]+beta,0,255)
        save_path = tarDir+name.split('.')[0]+'i.png'
        print(save_path)
        cv2.imwrite(save_path,new_image[:,:,[2,1,0]],[int(cv2.IMWRITE_PNG_COMPRESSION), 0])
#Decrease 
def Decrease(fileDir,tarDir,rate):
    pathDir = os.listdir(fileDir);
    filenumber = len(pathDir)
    picknumber=int(filenumber*rate) #按照rate比例从文件夹中取一定数量图片
    sample = random.sample(pathDir, picknumber)  #随机选取picknumber数量的样本图片
    for name in sample:
        img_path = fileDir+name
        image = imread(img_path)
        new_image = np.zeros(image.shape, image.dtype)
        alpha = 0.9 # Simple contrast control
        beta = 0.8  # Simple brightness control
        new_image[:,:,0] = np.clip(alpha*image[:,:,0]+beta,0,255)
        new_image[:,:,1] = np.clip(alpha*image[:,:,1]+beta,0,255)
        new_image[:,:,2] = np.clip(alpha*image[:,:,2]+beta,0,255)
        save_path = tarDir+name.split('.')[0]+'d.png'
        print(save_path)
        cv2.imwrite(save_path,new_image[:,:,[2,1,0]],[int(cv2.IMWRITE_PNG_COMPRESSION), 0])
# roration  
def rotation(fileDir,tarDir,rate):
    pathDir = os.listdir(fileDir);
    filenumber = len(pathDir)
    picknumber=int(filenumber*rate) #按照rate比例从文件夹中取一定数量图片
    sample = random.sample(pathDir, picknumber)  #随机选取picknumber数量的样本图片
    for name in sample:
        img_path = fileDir+name
#         image = imread(img_path)
        print(img_path)
        img = Image.open(img_path)
        angle = np.random.randint(10,25)
        rotation_img = img.rotate(angle) #旋转角度
        #保存图片
        save_path = tarDir+name.split('.')[0]+'r.png'
        print(save_path)
        rotation_img.save(save_path)
#solt noise
def solt(fileDir,tarDir,rate):
    pathDir = os.listdir(fileDir);
    filenumber = len(pathDir)
    picknumber=int(filenumber*rate) #按照rate比例从文件夹中取一定数量图片
    sample = random.sample(pathDir, picknumber)  #随机选取picknumber数量的样本图片
    for name in sample:
        img_path = fileDir+name
        image = imread(img_path)
        new_image = np.zeros(image.shape, image.dtype)
        alpha = 1.1 # Simple contrast control
        beta = 1.1  # Simple brightness control
        y = image[:,:,0];x = image[:,:,1];c = image[:,:,2]

        new_image[:,:,0] = np.clip(alpha*image[:,:,0]+beta,0,255)
        new_image[:,:,1] = np.clip(alpha*image[:,:,1]+beta,0,255)
        new_image[:,:,2] = np.clip(alpha*image[:,:,2]+beta,0,255)
        image = new_image
        #设置添加椒盐噪声的数目比例
        s_vs_p = 0.1
        #设置添加噪声图像像素的数目
        amount = 0.01
        noisy_img = np.copy(image)
        #添加salt噪声
        num_salt = np.ceil(amount * image.size * s_vs_p)
        #设置添加噪声的坐标位置
        coords = [np.random.randint(0,i - 1, int(num_salt)) for i in image.shape]
        noisy_img[coords] = 255
        #添加pepper噪声
        num_pepper = np.ceil(amount * image.size * (1. - s_vs_p))
        #设置添加噪声的坐标位置
        coords = [np.random.randint(0,i - 1, int(num_pepper)) for i in image.shape]
        noisy_img[coords] = 0
        #保存图片
        save_path = tarDir+name.split('.')[0]+'so.png'
        print(save_path)
        cv2.imwrite(save_path,new_image[:,:,[2,1,0]],[int(cv2.IMWRITE_PNG_COMPRESSION), 0])
        
yuan_train = "E:\\20-21_UAV_image\\WGSD\\CN-image\\"

save_train = "E:\\20-21_UAV_image\\WGSD\\test_save\\"


if __name__ == '__main__':
    for oneDir in os.listdir(yuan_train):#orignal 下的7个分类
        fileDir_train = yuan_train + oneDir + "/" 
        
        traDir = save_train + oneDir + "/"  # B的二级目录

        # 判断文件夹是否存在，不存在则创建
        if not os.path.exists(traDir):
            os.makedirs(traDir)
        sharpening(fileDir_train,sharpen,traDir,rate=0.1)
        Decrease(fileDir_train,traDir,rate=0.1)
        Increase(fileDir_train,traDir,rate=0.1)
        solt(fileDir_train,traDir,rate=0.1)
        rotation(fileDir_train,traDir,rate=0.1)
```
